```python
import pandas as pd

print(pd.Timestamp.now() )
print("-")

print("reading data file")
newdf = pd.DataFrame(pd.read_pickle("newdf.data"))
stagedf = pd.read_pickle("stagedf.data")

print("fixing bad values")
stagedf.fillna("", inplace=True)
stagedf = stagedf.replace(["Valores mínimos"], "")
stagedf = stagedf.replace(["Valores Mínimos"], "")
stagedf = stagedf.replace(["Percentual"], "")
stagedf = stagedf.replace(["Percentuais"], "")
stagedf = stagedf.replace(["Unnamed: 0"], "")
stagedf = stagedf.replace(["Unnamed: 1"], "")
stagedf = stagedf.replace(["Unnamed: 2"], "")
stagedf = stagedf.replace(["nan"], "")
stagedf = stagedf.replace(["-"], "")
    
    
print("handling data")
newdf = newdf.fillna("")
a, b, c, d, e, f, g, h = "","","","","","","",""
lastA, lastB, lastC = "","",""
for i, row in stagedf.iterrows():  
    a = stagedf.iloc[i, 0]
    if a == "" and b != "":
        a = lastA + ".x"
    else:
        lastA = a            
    b = stagedf.iloc[i, 1]  
    c = stagedf.iloc[i, 2]  
    d = stagedf.iloc[i, 3]     
    e =  stagedf.iloc[i, 4]   
    f = ""
    cds = a.split(".")
    if len(cds) == 1:  
        f = b
        lastB = b
    else:
        f = lastB                
    g = ""
    cds = a.split(".")
    if len(cds) == 2:
        if cds[1] == "1": 
            g = b
            lastC = b
    else:
        g = lastC
    h = ""    
    newrow = pd.DataFrame({'codigo':a,'atividade':b,'valor':c,'porcentagem':d,'sequencial':e,'materia':f,'classe':g,'tipo':h}, index=[0])
    newdf = pd.concat([newdf,newrow]).reset_index(drop=True)

print("saving finaldf dataframe")
newdf.to_pickle("finaldf.data")

print("-")
print(pd.Timestamp.now() )


pd.set_option("display.max_rows", None, "display.max_columns", None)
display(newdf)
```
