```python
import pandas as pd
import tabula
import os
from dotenv import load_dotenv
load_dotenv()
local = os.environ["originlocal"]

def getdf(pags,tab):
    tbs = tabula.read_pdf(local, pages=pags, lattice=True, multiple_tables=True, java_options="-Dfile.encoding=UTF8",)
    df = tbs[tab]
    
    c1 = df.columns.values[0]
    c2 = df.columns.values[1]
    c3 = df.columns.values[2]
    c4 = df.columns.values[3]
    df.columns.values[0] = "A"
    df.columns.values[1] = "B"
    df.columns.values[2] = "C"
    df.columns.values[3] = "D"
    new_row = pd.DataFrame({'A': c1, 'B': c2, 'C': c3, 'D': c4}, index=[0])
    df = pd.concat([new_row,df.loc[:]]).reset_index(drop=True)
    df = df.astype({"A": str})
    return df

print(pd.Timestamp.now() )
print("-")

print("creating target dataframe")
df = getdf("5", 0).copy()
dt = df.copy()
print(" - 01 ok")
df = getdf("5", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 02 ok")
df = getdf("5", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 03 ok")
df = getdf("6", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 04 ok")
df = getdf("6", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 04.1 ok")
df = getdf("6", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 04.2 ok")
df = getdf("6", 3).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 04.3 ok")
df = getdf("7", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 04.4 ok")
df = getdf("7", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 05 ok")
df = getdf("7", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 06 ok")
df = getdf("8", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 06.1 ok")
df = getdf("9", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 06.2 ok")
df = getdf("9", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 07 ok")
df = getdf("10", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 07.1 ok")
df = getdf("10", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 07.2 ok")
df = getdf("11", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 07.3 ok")
df = getdf("11", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 07.4 ok")
df = getdf("11", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 07.5 ok")
df = getdf("12", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 08 ok")
df = getdf("13", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 08.1 ok")
df = getdf("13", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 09 ok")
df = getdf("13", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 10 ok")
df = getdf("13", 3).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 10.1 ok")
df = getdf("14", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 10.2 ok")
df = getdf("14", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 11 ok")
df = getdf("14", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 12 ok")
df = getdf("14", 3).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 13 ok")
df = getdf("15", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 13.1 ok")
df = getdf("16", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 14 ok")
df = getdf("16", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 15 ok")
df = getdf("16", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 15.1 ok")
df = getdf("17", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 16 ok")
df = getdf("17", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 16.1 ok")
df = getdf("17", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 17 ok")
df = getdf("18", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 17.1 ok")
df = getdf("18", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 18 ok")
df = getdf("18", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 19 ok")
df = getdf("18", 3).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 20 ok")
df = getdf("19", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 20.1 ok")
df = getdf("19", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 21 ok")
df = getdf("19", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 22 ok")
df = getdf("19", 3).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 23 ok")
df = getdf("20", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 23.1 ok")
df = getdf("20", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 24 ok")
df = getdf("20", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 25 ok")
df = getdf("21", 0).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 26 ok")
df = getdf("21", 1).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 27 ok")
df = getdf("21", 2).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 28 ok")
df = getdf("21", 3).copy()
fr = [dt,df]
dt = pd.concat(fr)
print(" - 29 ok")     

print("handling dataframe")
dt = dt.reset_index(drop=True)
dt["sequencial"] = dt.index.astype(str).str.zfill(4)
    
print("saving pdf dataframe")
dt.to_pickle("pdfdf.data")

print("-")
print(pd.Timestamp.now() )


pd.set_option("display.max_rows", None, "display.max_columns", None)
display(dt)

```
